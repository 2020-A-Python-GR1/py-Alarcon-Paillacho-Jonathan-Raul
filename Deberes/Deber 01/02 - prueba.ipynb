```python
# Medicamentos
diccionario_medicamento_1 = {}

def crear_medicamento():
    nombre_far = input("¿En que farmacia va a estar disponible el medicamento? -> ")
    nombre_medic = input("Ingrese el nombre del medicamento: ")
    precio_medic = float(input(f"Ingrese el precio del medicamento {nombre_medic}: "))
    fecha_medic = input(f"Ingrese la fecha de fabricación del medicamento {nombre_medic}: ")
    unidades_medic = int(input(f"¿Cuántas unidades ingresan a la farmacia {nombre_far} del medicamento {nombre_medic}? -> "))
    ninios = input(f"\n¿El medicamento {nombre_medic} es apto para niños? -> ")
    if ninios == "Si":
        prevencion = True
    else:
        prevencion = False
    
    guardar_medicamento(nombre_far,nombre_medic,precio_medic,fecha_medic,unidades_medic,prevencion)
    

def guardar_medicamento(nombre_far,nombre_medic,precio_medic,fecha_medic,unidades_medic,prevencion):
    try:
        path = "./medicamento.txt"
        archivo_abierto = open(path,'a')
        archivo_abierto.write(f"{nombre_far},{nombre_medic},{precio_medic},{fecha_medic},{unidades_medic},{prevencion}\n")
        archivo_abierto.close()
    except Exception as Error:
        print("Error leyendo archivo")


        
def ver_medicamentos():
    try:
        path = "./medicamento.txt"
        archivo_abierto = open(path)
        medicamentos_lec = archivo_abierto.readlines()
        for lineas in medicamentos_lec:
            almacenar = lineas.split(',')
            diccionario_medicamento_2 = {
                f"{almacenar[1]}" : 
                [{"nombre_farmacia":f"{almacenar[0]}"},
                 {"precio":f"{almacenar[2]}"},
                 {"fecha":f"{almacenar[3]}"},
                 {"cantidad":f"{almacenar[4]}"},
                 {"menores":f"{almacenar[5]}"}]
            }
            diccionario_medicamento_1.update(diccionario_medicamento_2)
        archivo_abierto.close()
    except Exception as Error:
        print("Error leyendo archivo ver")

        
        
def buscar_medicamento(nombre = "todos"):
    ver_medicamentos()
    seguro = f"{nombre}" in diccionario_medicamento_1
    if nombre == "todas":
        print(diccionario_medicamento_1)
    elif seguro == True:
        print(diccionario_medicamento_1[f"{nombre}"])
    else:
        print("ATENCION: no existe la farmacia que está buscando!")
        


def actualizar_medicamento(nombre):
    ver_medicamentos()
    nombre_cambios = ["nombre farmacia","precio","fecha","cantidad","menores"]
    dic_actualizar = diccionario_medicamento_1.copy()
    seguro = f"{nombre}" in diccionario_medicamento_1
    if seguro == True:
        print(f"1. {nombre_cambios[0]}\n")
        print(f"2. {nombre_cambios[1]}\n")
        print(f"3. {nombre_cambios[2]}\n")
        print(f"4. {nombre_cambios[3]}\n")
        print(f"5. {nombre_cambios[4]}\n")
        eleccion = int(input("Elija el campo que desea actualizar -> "))
        cambio = input(f"Ingrese el nuevo valor para {nombre_cambios[eleccion-1]}: ")
        
        if eleccion == 1:
            dic_actualizar[f'{nombre}'][0]['nombre_farmacia'] = cambio
            
        elif eleccion == 2:
            dic_actualizar[f'{nombre}'][1]['precio'] = cambio
            
        elif eleccion == 3:
            dic_actualizar[f'{nombre}'][2]['fecha'] = cambio
            
        elif eleccion == 4:
            dic_actualizar[f'{nombre}'][3]['cantidad'] = cambio
            
        else:
            dic_actualizar[f'{nombre}'][4]['menores'] = cambio
            
        
        guardar_actualizacion_med(dic_actualizar)
    else:
        print("ATENCION: no existe la farmacia que desea actualizar!")
        

def guardar_actualizacion_med(nuevo_diccionario):
    #ver_medicamentos()
    try:
        path = "./medicamento.txt"
        archivo_abierto = open(path, 'w')
        archivo_abierto.write()
        archivo_abierto.close()
    except Exception as Error:
        print("Error leyendo archivo borrando")
    
    for llaves in nuevo_diccionario.keys():
        nombre = llaves
        nombre_farmacia = nuevo_diccionario[f'{nombre}'][0]['nombre_farmacia']
        precio = nuevo_diccionario[f'{nombre}'][1]['precio']
        fecha = nuevo_diccionario[f'{nombre}'][2]['fecha']
        cantidad = nuevo_diccionario[f'{nombre}'][3]['cantidad']
        menores = nuevo_diccionario[f'{nombre}'][4]['menores']
        try:
            path = "./medicamento.txt"
            archivo_abierto = open(path,'a')
            archivo_abierto.write(f"{nombre_farmacia},{nombre},{precio},{fecha},{cantidad},{menores}")
            archivo_abierto.close()
        except Exception as Error:
            print("Segundo error")
               

def eliminar_medicamento(nombre):
    ver_medicamentos()
    elim_med = diccionario_medicamento_1
    elim_med.pop(f'{nombre}')
    guardar_actualizacion_med(elim_med)       
        
#crear_medicamento()
#ver_medicamentos()
#diccionario_medicamento_1
#busqueda = input("Ingrese el nombre del medicamento que desea buscar o digite '"'todos'"' para mostrar todos los medicamentos: ")
#buscar_medicamento(busqueda)

#actualizacion_med = input("¿Qué Medicamento desea actualizar?: ")
#actualizar_medicamento(actualizacion_med)


#medi_suprimir = input("¿Qué Medicamento desea eliminar?: ")
#eliminar_medicamento(medi_suprimir)

#ver_medicamentos()
#diccionario_medicamento_1
```
