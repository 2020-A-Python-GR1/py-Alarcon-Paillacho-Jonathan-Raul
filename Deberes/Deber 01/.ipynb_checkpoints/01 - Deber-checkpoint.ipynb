```python
#Programa

def programa():
    
    def menu_inicial():
        print("¿Qué desea gestionar?")
        print("1. Farmacia")
        print("2. Medicamento")
        print("3. Salir")
        eleccion = int(input("Elija una opción -> "))
        if eleccion == 1:
            farmacia()
        elif eleccion == 2:
            medicamento()
        elif eleccion == 3:
            print("GRACIAS POR USAR NUESTRO SISTEMA")
        else:
            print("Ingresó una opción incorrecta")
        
    def farmacia():
        f = Farmacia()
        m = Medicamento()
        print("\n********************************************************\n")
        print("1. Crear una farmacia")
        print("2. Buscar una farmacia")
        print("3. Actualizar una farmacia")
        print("4. Eliminar una farmacia")
        eleccion = int(input("Elija una opción -> "))
        if eleccion == 1:
            f.crear_farmacia()
        elif eleccion == 2:
            busqueda = input("Ingrese el nombre de la farmacia que desea buscar o digite '"'todas'"' para mostrar todas las farmacias: ")
            f.buscar_farmacia(busqueda)
        elif eleccion == 3:
            actualizacion = input("¿Qué Farmacia desea actualizar?: ")
            f.actualizar_farmacia(actualizacion)
        elif eleccion == 4:
            farm_suprimir = input("¿Qué Farmacia desea eliminar?: ")
            f.eliminar_farmacia(farm_suprimir)
            m.eliminar_con_farmacia(farm_suprimir)
        else:
            print("Volver al menu inicial")
            menu_inicial()
    
    def medicamento():
        m = Medicamento()
        print("\n********************************************************\n")
        print("1. Crear un medicamento")
        print("2. Buscar un medicamento")
        print("3. Actualizar un medicamento")
        print("4. Eliminar un medicamento")
        eleccion = int(input("Elija una opción -> "))
        if eleccion == 1:
            m.crear_medicamento()
        elif eleccion == 2:
            busqueda = input("Ingrese el nombre del medicamento que desea buscar o digite '"'todos'"' para mostrar todos los medicamentos: ")
            m.buscar_medicamento(busqueda)
        elif eleccion == 3:
            actualizacion = input("¿Qué Medicamento desea actualizar?: ")
            m.actualizar_medicamento(actualizacion)
        elif eleccion == 4:
            med_suprimir = input("¿Qué Medicamento desea eliminar?: ")
            m.eliminar_medicamento(med_suprimir)
        else:
            print("Volver al menu inicial")
            menu_inicial()
            
    
    return menu_inicial()


programa()
```

```python
class Farmacia:
    
    diccionario_farmacia_1 = {}
   
    def __init__(self):
        print("Comienza el constructor de la farmacia")
    
    def crear_farmacia(self):
        medi = Medicamento()
        nombre_far = input("Ingrese el nombre de la farmacia: ")
        direccion_far = input(f"Ingrese la dirección de la farmacia {nombre_far}: ")
        telefono_far = int(input(f"Ingrese el numero de telefono de la farmacia {nombre_far}: "))
        compra_far = float(input("¿Cuánto es la mínima cantidad de compra? -> "))
        print("\nEn la siguiente pregunta ingrese Si o No")
        horario_far = input(f"\n¿La farmacia {nombre_far} atiende las 24 horas? -> ")
        if horario_far == "Si":
            atencion = True
        else:
            atencion = False
        
        
        print("\nEn la siguiente pregunta ingrese Si o No")
        agregar_med = input(f"Desea agregar un nuevo medicamento a la farmacia {nombre_far}? -> ")
        if agregar_med == "Si":
            medi.crear_medicamento(nombre_far)
        else:
            print("La farmacia fue creada con éxito")
            
        self.guardar_farmacia(nombre_far,direccion_far,telefono_far,compra_far,atencion)    
    
    def guardar_farmacia(self,nombre_far,direccion_far,telefono_far,compra_far,atencion):
        try:
            path = "./farmacia.txt"
            archivo_abierto = open(path,'a')
            archivo_abierto.write(f"{nombre_far},{direccion_far},{telefono_far},{compra_far},{atencion}\n")
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo")

    def ver_farmacias(self):
        try:
            path = "./farmacia.txt"
            archivo_abierto = open(path)
            leer_farmacias = archivo_abierto.readlines()
            for lineas in leer_farmacias:
                almacenar = lineas.split(',')
                diccionario_farmacia_2 = {
                    f"{almacenar[0]}" : 
                    [{"direccion":f"{almacenar[1]}"},
                     {"telefono":f"{almacenar[2]}"},
                     {"compra":f"{almacenar[3]}"},
                     {"atencion":f"{almacenar[4]}"}]
                }
                #self.diccionario_farmacia_1 = diccionario_farmacia_1
                self.diccionario_farmacia_1.update(diccionario_farmacia_2)
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo ver")


    def buscar_farmacia(self,nombre = "todas"):
        self.ver_farmacias()
        seguro = f"{nombre}" in self.diccionario_farmacia_1
        if nombre == "todas":
            print(self.diccionario_farmacia_1)
        elif seguro == True:
            print(self.diccionario_farmacia_1[f"{nombre}"])
        else:
            print("ATENCION: no existe la farmacia que está buscando!")

            
    def actualizar_farmacia(self,nombre):
        self.ver_farmacias()
        nombre_cambios = ["direccion","telefono","compra","atencion"]
        dic_actualizar = self.diccionario_farmacia_1.copy()
        seguro = f"{nombre}" in self.diccionario_farmacia_1
        if seguro == True:
            print(f"1. {nombre_cambios[0]}\n")
            print(f"2. {nombre_cambios[1]}\n")
            print(f"3. {nombre_cambios[2]}\n")
            print(f"4. {nombre_cambios[3]}\n")
            eleccion = int(input("Elija el campo que desea actualizar -> "))
            cambio = input(f"Ingrese el nuevo valor para {nombre_cambios[eleccion-1]}: ")

            if eleccion == 1:
                #print("Entra 1")
                dic_actualizar[f'{nombre}'][0]['direccion'] = cambio
                
            elif eleccion == 2:
                dic_actualizar[f'{nombre}'][1]['telefono'] = cambio
                
            elif eleccion == 3:
                dic_actualizar[f'{nombre}'][2]['compra'] = cambio
                
            else:
                dic_actualizar[f'{nombre}'][3]['atencion'] = cambio
                

            
            self.guardar_actualizacion(dic_actualizar)
        else:
            print("ATENCION: no existe la farmacia que desea actualizar!")

    

    def guardar_actualizacion(self,nuevo_diccionario):
        
        try:
            path = "./farmacia.txt"
            archivo_abierto = open(path, 'w')
            archivo_abierto.write()
            archivo_abierto.close()
        except Exception as Error:
            print("Los campos de la farmacia se actualizaron con éxito")

        for llaves in nuevo_diccionario.keys():
            nombre = llaves
            #print(nombre)
            direccion = nuevo_diccionario[f'{nombre}'][0]['direccion']
            telefono = nuevo_diccionario[f'{nombre}'][1]['telefono']
            compra = nuevo_diccionario[f'{nombre}'][2]['compra']
            atencion = nuevo_diccionario[f'{nombre}'][3]['atencion']
            try:
                path = "./farmacia.txt"
                archivo_abierto = open(path,'a')
                archivo_abierto.write(f"{nombre},{direccion},{telefono},{compra},{atencion}")
                archivo_abierto.close()
            except Exception as Error:
                print("Segundo error")
                    

    def eliminar_farmacia(self,nombre):
        self.ver_farmacias()
        elim_far = self.diccionario_farmacia_1
        elim_far.pop(f'{nombre}')
        self.guardar_actualizacion(elim_far)

#f = Farmacia()
#f.crear_farmacia()
```

```python
# Medicamentos
class Medicamento:
    
    diccionario_medicamento_1 = {}
    diccionario_farmacia = {}
    diccionario_farmacia_elim = {}
    def __init__(self):
        self.diccionario_medicamento_1
        self.diccionario_farmacia
        self.diccionario_farmacia_elim
        print("Empieza el constructor de la clase Medicamento")

    def crear_medicamento(self, nombre_farm = "nada"):
        self.medicamento_en_farmacia()
        if nombre_farm == "nada":
            nombre_far = input("¿En que farmacia va a estar disponible el medicamento? -> ")
            if nombre_far in self.diccionario_farmacia:
                nombre_medic = input("Ingrese el nombre del medicamento: ")
                precio_medic = float(input(f"Ingrese el precio del medicamento {nombre_medic}: "))
                fecha_medic = input(f"Ingrese la fecha de fabricación del medicamento {nombre_medic}: ")
                unidades_medic = int(input(f"¿Cuántas unidades ingresan a la farmacia {nombre_far} del medicamento {nombre_medic}? -> "))
                print("\nEn la siguiente pregunta ingrese Si o No")
                ninios = input(f"\n¿El medicamento {nombre_medic} es apto para niños? -> ")
                if ninios == "Si":
                    prevencion = True
                else:
                    prevencion = False

                self.guardar_medicamento(nombre_far,nombre_medic,precio_medic,fecha_medic,unidades_medic,prevencion)
            else:
                print("El nombre de farmacia que ingresó no existe en la base de datos")
            
            #self.medicamento_en_farmacia()
            #self.guardar_medicamento_en_farmacia(self.diccionario_farmacia,nombre_medic,nombre_far)
        else:
            print(nombre_farm)
            nombre_medic = input("Ingrese el nombre del medicamento: ")
            precio_medic = float(input(f"Ingrese el precio del medicamento {nombre_medic}: "))
            fecha_medic = input(f"Ingrese la fecha de fabricación del medicamento {nombre_medic}: ")
            unidades_medic = int(input(f"¿Cuántas unidades ingresan a la farmacia {nombre_farm} del medicamento {nombre_medic}? -> "))
            print("\nEn la siguiente pregunta ingrese Si o No")
            ninios = input(f"\n¿El medicamento {nombre_medic} es apto para niños? -> ")
            if ninios == "Si":
                prevencion = True
            else:
                prevencion = False

            self.guardar_medicamento(nombre_farm,nombre_medic,precio_medic,fecha_medic,unidades_medic,prevencion)
            #self.medicamento_en_farmacia()
            #self.guardar_medicamento_en_farmacia(self.diccionario_farmacia,nombre_medic,nombre_far)
            
        
        


    def guardar_medicamento(self,nombre_far,nombre_medic,precio_medic,fecha_medic,unidades_medic,prevencion):
        try:
            path = "./medicamento2.txt"
            archivo_abierto = open(path,'a')
            archivo_abierto.write(f"{nombre_far},{nombre_medic},{precio_medic},{fecha_medic},{unidades_medic},{prevencion}\n")
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo")

    def medicamento_en_farmacia(self):
        try:
            path = "./farmacia.txt"
            archivo_abierto = open(path)
            leer_farmacias = archivo_abierto.readlines()
            for lineas in leer_farmacias:
                almacenar = lineas.split(',')
                diccionario_farmacia_2 = {
                    f"{almacenar[0]}" : 
                    [{"direccion":f"{almacenar[1]}"},
                     {"telefono":f"{almacenar[2]}"},
                     {"compra":f"{almacenar[3]}"},
                     {"atencion":f"{almacenar[4]}"}]
                }
                
                self.diccionario_farmacia.update(diccionario_farmacia_2)
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo ver")
        
    def guardar_medicamento_en_farmacia(self,nuevo_diccionario,nombre_med,nombre_farm):
        try:
            path = "./farmacia.txt"
            archivo_abierto = open(path, 'w')
            archivo_abierto.write()
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo borrando")

        for llaves in nuevo_diccionario.keys():
            nombre = llaves
            if nombre == nombre_farm:
                direccion = nuevo_diccionario[f'{nombre}'][0]['direccion']
                telefono = nuevo_diccionario[f'{nombre}'][1]['telefono']
                compra = nuevo_diccionario[f'{nombre}'][2]['compra']
                atencion = nuevo_diccionario[f'{nombre}'][3]['atencion']
                try:
                    path = "./farmacia.txt"
                    archivo_abierto = open(path,'a')
                    archivo_abierto.write(f"{nombre},{direccion},{telefono},{compra},{atencion},{nombre_med}")
                    archivo_abierto.close()
                except Exception as Error:
                    print("Segundo error")
            else:
                direccion = nuevo_diccionario[f'{nombre}'][0]['direccion']
                telefono = nuevo_diccionario[f'{nombre}'][1]['telefono']
                compra = nuevo_diccionario[f'{nombre}'][2]['compra']
                atencion = nuevo_diccionario[f'{nombre}'][3]['atencion']
                try:
                    path = "./farmacia.txt"
                    archivo_abierto = open(path,'a')
                    archivo_abierto.write(f"{nombre},{direccion},{telefono},{compra},{atencion}")
                    archivo_abierto.close()
                except Exception as Error:
                    print("Segundo error")
            
                 
        
    def buscar_medicamento(self,nombre = "todos"):
        self.ver_medicamentos()
        seguro = f"{nombre}" in self.diccionario_medicamento_1
        if nombre == "todas":
            print(self.diccionario_medicamento_1)
        elif seguro == True:
            print(self.diccionario_medicamento_1[f"{nombre}"])
        else:
            print("ATENCION: no existe la farmacia que está buscando!")

    def ver_medicamentos(self):
        try:
            path = "./medicamento2.txt"
            archivo_abierto = open(path)
            medicamentos_lec = archivo_abierto.readlines()
            for lineas in medicamentos_lec:
                almacenar = lineas.split(',')
                diccionario_medicamento_2 = {
                    f"{almacenar[1]}" : 
                    [{"nombre_farmacia":f"{almacenar[0]}"},
                     {"precio":f"{almacenar[2]}"},
                     {"fecha":f"{almacenar[3]}"},
                     {"cantidad":f"{almacenar[4]}"},
                     {"menores":f"{almacenar[5]}"}]
                }
                self.diccionario_medicamento_1.update(diccionario_medicamento_2)
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo ver")

    def actualizar_medicamento(self,nombre):
        self.ver_medicamentos()
        nombre_cambios = ["nombre farmacia","precio","fecha","cantidad","menores"]
        dic_actualizar = self.diccionario_medicamento_1.copy()
        seguro = f"{nombre}" in self.diccionario_medicamento_1
        if seguro == True:
            print(f"1. {nombre_cambios[0]}\n")
            print(f"2. {nombre_cambios[1]}\n")
            print(f"3. {nombre_cambios[2]}\n")
            print(f"4. {nombre_cambios[3]}\n")
            print(f"5. {nombre_cambios[4]}\n")
            eleccion = int(input("Elija el campo que desea actualizar -> "))
            cambio = input(f"Ingrese el nuevo valor para {nombre_cambios[eleccion-1]}: ")

            if eleccion == 1:
                dic_actualizar[f'{nombre}'][0]['nombre_farmacia'] = cambio

            elif eleccion == 2:
                dic_actualizar[f'{nombre}'][1]['precio'] = cambio

            elif eleccion == 3:
                dic_actualizar[f'{nombre}'][2]['fecha'] = cambio

            elif eleccion == 4:
                dic_actualizar[f'{nombre}'][3]['cantidad'] = cambio

            else:
                dic_actualizar[f'{nombre}'][4]['menores'] = cambio


            self.guardar_actualizacion_med(dic_actualizar)
        else:
            print("ATENCION: no existe la farmacia que desea actualizar!")


    def guardar_actualizacion_med(self,nuevo_diccionario):
        #ver_medicamentos()
        try:
            path = "./medicamento2.txt"
            archivo_abierto = open(path, 'w')
            archivo_abierto.write()
            archivo_abierto.close()
        except Exception as Error:
            print("Los cambios se guardaron correctamente")

        for llaves in nuevo_diccionario.keys():
            nombre = llaves
            nombre_farmacia = nuevo_diccionario[f'{nombre}'][0]['nombre_farmacia']
            precio = nuevo_diccionario[f'{nombre}'][1]['precio']
            fecha = nuevo_diccionario[f'{nombre}'][2]['fecha']
            cantidad = nuevo_diccionario[f'{nombre}'][3]['cantidad']
            menores = nuevo_diccionario[f'{nombre}'][4]['menores']
            try:
                path = "./medicamento2.txt"
                archivo_abierto = open(path,'a')
                archivo_abierto.write(f"{nombre_farmacia},{nombre},{precio},{fecha},{cantidad},{menores}")
                archivo_abierto.close()
            except Exception as Error:
                print("Segundo error")


    def eliminar_medicamento(self,nombre):
        self.ver_medicamentos()
        elim_med = self.diccionario_medicamento_1
        elim_med.pop(f'{nombre}')
        self.guardar_actualizacion_med(elim_med)
        
        
    def eliminar_con_farmacia(self,nombre_farma):
        diccionario_farmacia_elim = {}
        try:
            path = "./medicamento2.txt"
            archivo_abierto = open(path)
            leer_farmacias = archivo_abierto.readlines()
            for lineas in leer_farmacias:
                almacenar = lineas.split(',')
                diccionario_farmacia_2 = {
                    f"{almacenar[0]}" : 
                    [{"nombre_med":f"{almacenar[1]}"},
                     {"precio":f"{almacenar[2]}"},
                     {"fecha":f"{almacenar[3]}"},
                     {"cantidad":f"{almacenar[4]}"},
                     {"menores":f"{almacenar[5]}"}]
                }
                
                diccionario_farmacia_elim.update(diccionario_farmacia_2)
            archivo_abierto.close()
        except Exception as Error:
            print("Error leyendo archivo ver")
        comprobar = nombre_farma in diccionario_farmacia_elim
        if comprobar == True:
            diccionario_farmacia_elim.pop(f'{nombre_farma}')
            self.guardar_elim(diccionario_farmacia_elim)
        else:
            print("No se encuentra")
            
    
    def guardar_elim(self,nuevo_diccionario):
        #print("Entra en la funcion guardar")
        try:
            path = "./medicamento2.txt"
            archivo_abierto = open(path, 'w')
            archivo_abierto.write()
            archivo_abierto.close()
        except Exception as Error:
            print("Los campos se actualizaron con éxito, Funcionó correctamente")

        for llaves in nuevo_diccionario.keys():
            nombre_farmacia = llaves
            nombre_med = nuevo_diccionario[f'{nombre_farmacia}'][0]['nombre_med']
            precio = nuevo_diccionario[f'{nombre_farmacia}'][1]['precio']
            fecha = nuevo_diccionario[f'{nombre_farmacia}'][2]['fecha']
            cantidad = nuevo_diccionario[f'{nombre_farmacia}'][3]['cantidad']
            menores = nuevo_diccionario[f'{nombre_farmacia}'][4]['menores']
            try:
                path = "./medicamento2.txt"
                archivo_abierto = open(path,'a')
                archivo_abierto.write(f"{nombre_farmacia},{nombre_med},{precio},{fecha},{cantidad},{menores}")
                archivo_abierto.close()
            except Exception as Error:
                print("Segundo error")

```
